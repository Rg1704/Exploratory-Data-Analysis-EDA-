```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import Csv File
df = pd.read_csv('train.csv')
```

```python
#sum up Null values Column Wise
df.isnull().sum()
```

```python
#give total null values
df.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.isnull())
```

```python
# Convert all null values to percentage
null_percentage = df.isnull().sum()/df.shape[0]*100
null_percentage
```

```python
null_percentage > 15
```

```python
#Here we will drop null values greater than 15%
drop_columns = null_percentage[null_percentage > 15].keys()
drop_columns

```

```python
#drop columns null values >15 and store in another dataframe variable.
df1_drop_clmns = df.drop(columns = drop_columns)
```

```python
# columns reduced from 81 to 75
df1_drop_clmns.shape
```

```python
#Again Checking null values
df1_drop_clmns.isnull().sum()
```

```python
#heat map after removing null values columns >17%
plt.figure(figsize=(16,9))
sns.heatmap(df1_drop_clmns.isnull())
```

```python
#drop all the rows with null values
df1_drop_rows = df1_drop_clmns.dropna()
```

```python
#After deleting null values rows reduced from 1460 -> 1338
df1_drop_rows.shape
```

```python
#plotting heat map after removing all the null values
plt.figure(figsize=(16,9))
sns.heatmap(df1_drop_rows.isnull())
```

```python
#check column wise null values in data after cleaning 
df1_drop_rows.isnull().sum()
```

```python
#check total null values in data after cleaning 
df1_drop_rows.isnull().sum().sum()
```
