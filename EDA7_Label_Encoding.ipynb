#Label encoding we give labels to the variables , that is we convert categorical variables into numerical variables by giving labelling them orders.
#ex:- (A,B,C)->(3,2,1) A as higher,(Excellent, Good, Bad)->(1,2,3) Excellent as higher..... (male, female)->(0,1) or(1,2) anything.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("train_house_price_prediction.csv")
df.head()
```

```python
df.info()
```

```python
df.select_dtypes(include="object").columns
```

#lets take random two columns 'BldgType','KitchenQual'

```python
df2=df[["KitchenQual","BldgType"]] #now create new dataframe with the selected two feature
df2
```

```python
le=LabelEncoder() #create an object 
```

```python
le.fit_transform(df2["BldgType"]) #here we got an array we will convert it into dataframe.
```

```python
df2["BldgType_L_enc"]=le.fit_transform(df2["BldgType"])
df2
```

```python
df["BldgType"].value_counts()#it will give label  alphabatically. so 1Fam=0,2fmCon=1,Duplex=2,Twnhs=3,TwnhsE=4
```

```python
df["KitchenQual"].value_counts() #now here 4 classe.Ex=excellent, GD=good, TA=typical avg,FA=Fair
```

# Ex=excellent> GD=good> TA=typical avg>FA=Fair

```python
"""# KitchenQual: Kitchen quality

       Ex	Excellent
       Gd	Good
       TA	Typical/Average
       Fa	Fair
"""
```

#Label Encoding

```python
order_Label = {"Ex":4,"Gd":3,"TA":2,"Fa":1}
```

```python
df2["KitchenQual_org_enc"] = df2["KitchenQual"].map(order_Label)
```
